```python
import pandas as pd

df = pd.read_csv('dataClean.csv')
df
```

```python
df[df['Category'] == 'Investment']
```

```python
df['Date'] = pd.to_datetime(df['Date'])  # Convert the 'Date' column to datetime format

# Break the 'Date' column into year, month, day, day of week features
df['year'] = df['Date'].dt.year
df['month'] = df['Date'].dt.month
df['day'] = df['Date'].dt.day
df['dayofweek'] = df['Date'].dt.dayofweek  #(0 = Monday, 6 = Sunday)
```

```python
df['year'].value_counts().sort_index(ascending=True)
```

```python
df['month'].value_counts().sort_index(ascending=True)
```

```python
# check day value sort descending
df['day'].value_counts().sort_index(ascending=True)
```

```python
df['dayofweek'].value_counts().sort_index(ascending=True)
```
